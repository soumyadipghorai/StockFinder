```python
import pandas as pd 
import plotly.express as px
```

```python
df = pd.read_csv('../dump/9-08-2024-TO-9-08-2025-CUMMINSIND-ALL-N.csv')
df["date"] = pd.to_datetime(df["Date  "])
df.set_index("date", inplace=True)
df['close'] = df['Close Price  '].astype(str)
df['close'] = df['close'].str.replace(",", "").astype(float)
df['close'] = df['close'].fillna(0)
df["EMA_50"] = df["close"].ewm(span=50, adjust=False).mean()
df["EMA_100"] = df["close"].ewm(span=100, adjust=False).mean()
df['diff'] = df['EMA_50'] - df['EMA_100']
fig = px.line(
    df,
    x=df.index, 
    y=["EMA_50", "EMA_100"]
)

fig.show()
df = df.tail(50)
arr = []
for i in range(len(df)) : 
    arr.append(True if df['diff'].iloc[i] > 0 else False) 

change_counter = 0 
curr, last_change_detected = arr[0], -1
for i in range(1, len(arr)) : 
    if arr[i] != curr : 
        change_counter += 1 
        curr = arr[i]
        last_change_detected = i

print(change_counter, len(arr) - last_change_detected)
print(arr)
```

```python
import os 
import json
csv_files = [f.split('-')[-3] for f in os.listdir('../dump') if f.endswith('.csv')]
unique_files = {}
for company in csv_files : 
    if company not in unique_files : 
        unique_files[company] = 0
    unique_files[company] += 1 

with open("../data/all_company.json", "r") as json_file:
    current_company = json.load(json_file)

missing_companies = []
for company in current_company :
    if company not in unique_files :
        missing_companies.append(company) 

print(missing_companies, len(missing_companies))
```

```python
import os 
import pandas as pd
import json
data = {}
for folder in os.listdir('../data'):
    folder_path = os.path.join('../data', folder)

    # Ensure it's a folder
    if os.path.isdir(folder_path):
        csv_path = os.path.join(folder_path, "company_info.csv")
        if os.path.exists(csv_path):
            try:
                # Read CSV into DataFrame
                df = pd.read_csv(csv_path)

                # Convert DataFrame to list of dicts (records)
                df = df.where(pd.notna(df), None)
                value_list = df.to_dict(orient="records")
                data[folder] = value_list
            except Exception as e:
                print(f"Error reading {csv_path}: {e}")
```

```python
with open('../data/current_trend_info.json', "w", encoding="utf-8") as f:
    json.dump(data, f, indent=4)
```
